```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the dataset
data = pd.read_csv('stock_prices.csv')
```

```python
data.head()
```

```python
# Convert the date column to datetime object
data['date'] = pd.to_datetime(data['date'])
```

```python
# Set the date column as the index
data.set_index('date', inplace=True)
```

```python
# Calculate the rolling 30-day moving average
rolling_avg = data['closing_price'].rolling(window=30).mean()
```

```python
rolling_avg
```

```python
# Plot the closing price and rolling average
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['closing_price'], label='Closing Price', color='blue')
plt.plot(data.index, rolling_avg, label='30-Day Rolling Avg', color='red')
plt.title('Stock Prices and 30-Day Rolling Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
```
